# Handling Categorical datas

**Approach 1 --> Label Encoding**

There are two types of categorical data
1) Ordinal Categorical data  ----->  order of categories

2) Non Ordinal Categorical data

```python
import pandas as pd

cricket = {'Team': ['India', 'India', 'Australia', 'Australia', 'SA',
   'SA', 'SA', 'SA', 'NZ', 'NZ', 'NZ', 'India'],
   'Rank': [2, 3, 1,2, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,801,891,815,776,784,834,824,758,691,883,782]}
df = pd.DataFrame(cricket)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
label_encoded_data=lb.fit_transform(df['Team'])
label_encoded_data=pd.DataFrame(data=label_encoded_data,columns=['Team'])
label_encoded_data
```

```python
print(df['Team'].unique())
print(label_encoded_data['Team'].unique())
```

**Approach 2 --> One Hot Encoding**

binary labelling 1 in places where the value is present else 0 should be present

```python
one_hot_encoding=pd.get_dummies(df,columns=['Team'])
```
